```python
for i in df.iterrows():
    if i[1][0] == "Metagraph":
        df.loc[i[0], 'x'] = np.random.randint(-100, 100)
        df.loc[i[0], 'y'] = np.random.randint(-100, 100)
        df.loc[i[0], 'z'] = np.random.randint(-100, 100) - 200

        if df.loc[i[0], 2] >13:
            df.loc[i[0], 2] = 30
        else:
            df.loc[i[0], 2] = 2

    elif any(df[1] == i[1][0] ):
        theta = np.random.randint(0, int(200*np.pi))/100
        phi = np.random.randint(0, int(200*np.pi))/100

        cos_phi = np.cos(phi)
        sin_phi = np.sin(phi)
        cos_theta = np.cos(theta)
        sin_theta = np.sin(theta)
        
        x = 10*np.outer(sin_phi, cos_theta)
        y = 10*np.outer(sin_phi, sin_theta)
        z = 10*np.outer(cos_phi, np.ones_like(theta))

        df.loc[i[0], 'x'] = df[df[1] == i[1][0]].x.values[0] + x
        df.loc[i[0], 'y'] = df[df[1] == i[1][0]].y.values[0] + y
        df.loc[i[0], 'z'] = df[df[1] == i[1][0]].z.values[0] + z

        if df.loc[i[0], 2] >13:
            df.loc[i[0], 2] = 30
        else:
            df.loc[i[0], 2] = 2
        
df
```
